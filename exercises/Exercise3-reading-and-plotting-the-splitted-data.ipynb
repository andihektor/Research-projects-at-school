```python
import pandas as pd # powerful Python data analysis toolkit
import numpy as np # numberical calculations
import matplotlib as plt # figures
import matplotlib.pyplot as plt # for a pythnon 3 bug
import astropy as ast # tools for astromomy
from astropy import units as u
from astropy.coordinates import SkyCoord
```

```python
data1 = pd.read_pickle("../data/subdata-0.pkl.gz", compression="gzip")
```

```python
# Pandas has a method for direct plotting
data1.plot('l','b', markersize=1, marker='o',linestyle='')
plt.grid(True)
plt.show()
```

```python
# Conversion functions from degrees to radians for the Aitoff projection
def prep_l(data):
    plot_l = lambda l: np.radians(l) if l < 180 else np.radians(l-360)
    return list(map(plot_l, data['l'].values))
def prep_b(data):
    plot_b = lambda b: np.radians(b)
    return list(map(plot_b, data['b'].values))
```

```python
# So the figure in the Aitoff
nnPlot = 100000
plt.figure(figsize=(8,4.2))
plt.subplot(111, projection="aitoff")
plt.grid(True)
plt.plot(prep_l(data1.loc[:nnPlot]), prep_b(data1.loc[:nnPlot]), '.', markersize=1, alpha=1)
plt.show()
```

```python
# Conversions for the degrees for histrogramming
def prep_l_deg(data):
    plot_l = lambda l: l if l < 180 else l-360
    return list(map(plot_l, data['l'].values))
def prep_b_deg(data):
    plot_b = lambda b: b
    return list(map(plot_b, data['b'].values))
```

```python
from matplotlib.colors import LogNorm
import pylab as plt
```

```python
# Histogramming
plt.figure(figsize=(8,4.2))
plt.hist2d(prep_l_deg(data1), prep_b_deg(data1), [100,50], norm=LogNorm())
plt.grid(True)
plt.colorbar()
plt.show()
```
